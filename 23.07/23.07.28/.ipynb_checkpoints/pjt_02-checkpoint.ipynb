```python
# 패키지 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# A. 데이터 전처리 - 데이터 읽어오기

```python
# CSV 파일 경로
csv_path = "data/NLFX.csv"

# CSV 파일 읽어오기
df = pd.read_csv(csv_path, usecols = range(0, 5))

# DataFrame 출력
df
```

# B. 데이터 전처리 - 2021년 이후의 종가 데이터 출력하기

```python
# 데이터 타입 변경
df["Date"] = pd.to_datetime(df["Date"])

# 2021년 이후 데이터 필터링
df_after_2021 = df[df["Date"].dt.year >= 2021]

# 그래프 그리기
plt.plot(df_after_2021["Date"], df_after_2021["Close"])

# 그래프 제목 설정
plt.title("NFLX Close Price")

# x축 레이블 설정
plt.xlabel("Date")

# x축 회전
plt.xticks(rotation = 45)

# y축 레이블 설정
plt.ylabel("Close Price")

# 그래프 표시
plt.show()
```

# C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기

```python
# 2021년 이후 최고 종가
max_price = df_after_2021["Close"].max()

# 2021년 이후 최저 종가
min_price = df_after_2021["Close"].min()

# 출력
print("최고 종가 :", max_price)
print("최저 종가 :", min_price)
```

# D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기

```python
# CSV 파일 읽어오기
df_new = pd.read_csv(csv_path, usecols = range(0, 5))

# 데이터 타입 변경
df_new["Date"] = pd.to_datetime(df_new["Date"])

# 2021년 이후 데이터 필터링
df_new_after_2021 = df_new[df_new["Date"].dt.year >= 2021]

# DataFrame 복사
df_new_after_2021_copy = df_new_after_2021.copy()

# 월 정보 새로운 컬럼으로 추가
df_new_after_2021_copy["Month"] = df_new_after_2021_copy["Date"].dt.strftime("%y-%m")

# 월 별로 그룹화하여 평균 종가 계산
average_close_price = df_new_after_2021_copy.groupby("Month")["Close"].mean()

# 새로운 DataFrame 만들기
df_monthly = pd.DataFrame({"Date" : average_close_price.index, "Average Close Price" : average_close_price.values})

# 그래프 그리기
plt.plot(df_monthly["Date"], df_monthly["Average Close Price"])

# 그래프 제목 설정
plt.title("Monthly Average Close Price")

# x축 레이블 설정
plt.xlabel("Date")

# x축 회전
plt.xticks(rotation = 45)

# y축 레이블 설정
plt.ylabel("Average Close Price")

# 그래프 표시
plt.show()
```

# E. 데이터 시각화 - 2022년 이후 최고, 최저, 종가 시각화하기

```python
# 2022년 이후 데이터 필터링
df_after_2022 = df[df["Date"].dt.year >= 2022]

# 그래프 그리기
plt.plot(df_after_2022["Date"], df_after_2022["High"], label = "High")
plt.plot(df_after_2022["Date"], df_after_2022["Low"], label = "Low")
plt.plot(df_after_2022["Date"], df_after_2022["Close"], label = "Close")

# 그래프 제목 설정
plt.title("High, Low, and Close Price since January 2022")

# x축 레이블 설정
plt.xlabel("Date")

# x축 회전
plt.xticks(rotation = 45)

# y축 레이블 설정
plt.ylabel("Price")

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
